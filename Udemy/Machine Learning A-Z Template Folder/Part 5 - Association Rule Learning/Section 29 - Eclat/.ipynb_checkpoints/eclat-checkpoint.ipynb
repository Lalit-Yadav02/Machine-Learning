```python
transactions = [[str(df.values[i, j]) for j in range(df.shape[1]) if str(df.values[i, j]) != 'nan'] for i in range(df.shape[0])]
```

```python
items = list()
[[items.append(x) for x in t if not x in items] for t in transactions];
```

```python
eclat = list()
[[eclat.append([(items[i], items[j]), 0]) for j in range(i+1, len(items))] for i in range(len(items))];
```

```python
for p in eclat:
    for t in transactions:
        if (p[0][0] in t) and (p[0][1] in t):
            p[1] += 1
    p[1] /= len(transactions)
```

```python
data = pd.DataFrame(eclat, columns = ['rule', 'support']).sort_values(by = 'support', ascending = False)
```
