```python
sigmoid = lambda z: (1/ (1 + np.exp(-z)))
```

```python
sigmoid_gradient = lambda z: sigmoid(z) * (1 - sigmoid(z))
```

```python
input_layer_size  = 400
hidden_layer_size = 25
num_labels = 10
```

```python
def nn_cost_function(nn_params, input_layer_size, hidden_layer_size, num_labels, X, y, lmda=0.0):
    theta1 = nn_params[: hidden_layer_size * (input_layer_size + 1)], (hidden_layer_size, (input_layer_size+1))
    theta2 = nn_params[(hidden_layer_size * (input_layer_size + 1)):].reshape(num_labels, (hidden_layer_size+1))
    m, n = X.shape
    
    a1 = X
    
    z2 = theta1.dot(a1.T)
    a2 = sigmoid(z2)
    a2 = np.r_[ np.ones((1, z2.shape[1])), a2]
    
    z3 = theta2.dot(a2)
    a3 = sigmoid(z3)
    
    Y = pd.get_dummies(y.ravel()).as_matrix()
    
    J = (-1/m) * np.trace(np.log(a3).dot(Y) + np.log(1 - a3).dot((1 - Y))) + \
        (lmda/ (2*m)) * (np.trace((theta1[:, 1:]).T.dot(theta1[:, 1:]))) + (np.trace((theta2[:, 1:]).T.dot(theta2[:, 1:])))
    
    delta3 = a3.T - Y
    delta2 = theta2[:, 1:].T.dot(delta3.T) * sigmoid_gradient(z2)
    
    D1 = delta2.dot(a1)
    D2 = delta3.T.dot(a2.T)
    
    theta1_reg = np.c_[np.ones((theta1.shape[0],1)),theta1[:,1:]]
    theta2_reg = np.c_[np.ones((theta2.shape[0],1)),theta2[:,1:]]
    
    theta1_grad = D1/m + (theta1_reg * lmda)/m 
    theta2_grad= D2/m + (theta2_reg * lmda)/m
    
    grad = np.r_[theta1_grad.ravel(), theta2_grad.ravel()]
    
    return (J, grad)
```

```python
lmda = 0
print(nn_cost_function(nn_params, input_layer_size, hidden_layer_size, num_labels, X, y, lmda))
```
