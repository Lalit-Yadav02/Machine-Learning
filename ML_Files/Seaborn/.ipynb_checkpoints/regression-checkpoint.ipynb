```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm', markers=['o', 'v'], scatter_kws={'s': 100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=tips, palette='coolwarm')
```

```python
sns.lmplot(x="total_bill", y="tip", col="day", data=tips, hue='sex', palette='coolwarm', aspect=0.6, size=8)
```
