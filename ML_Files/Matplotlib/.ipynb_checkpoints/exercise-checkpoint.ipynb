```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, 'b')
axes.set_title('Title')
axes.set_xlabel('X')
axes.set_ylabel('Y')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes = fig.add_axes([0.2,0.5,.2,.2])
```

```python
fig = plt.figure()
axes_1 = fig.add_axes([0, 0, 1, 1])
axes_2 = fig.add_axes([0.2,0.5,.2,.2])
axes_1.plot(x, y, 'r')
axes_1.set_title('Title')
axes_1.set_xlabel('X')
axes_1.set_ylabel('Y')
axes_2.plot(y, x, 'r')
axes_2.set_title('Title')
axes_2.set_xlabel('X')
axes_2.set_ylabel('Y')
```

```python
fig = plt.figure()
axes_1 = fig.add_axes([0, 0, 1, 1])
axes_2 = fig.add_axes([0.2,0.5,.4,.4])
axes_1.plot(x, z, 'b')
axes_1.set_title('Title')
axes_1.set_xlabel('X')
axes_1.set_ylabel('Y')
axes_2.plot(x, y, 'r')
axes_2.set_title('Title')
axes_2.set_xlabel('X')
axes_2.set_ylabel('Y')
axes_2.set_xlim(20, 22)
axes_2.set_ylim(30, 50)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, ls='--', c='b', lw=3)
axes[1].plot(x, z, 'r', lw=4)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 2))
axes[0].plot(x, y, ls='--', c='b', lw=3)
axes[1].plot(x, z, 'r', lw=4)
```
