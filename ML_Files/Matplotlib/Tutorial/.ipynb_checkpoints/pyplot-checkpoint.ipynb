```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])  # x and y are same so, x = 0-1 and y = 1-4
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) # plotting x vs y
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # plotting with red circle
```

```python
import numpy as np
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data) # data is like dictonary and other parma are key of data
plt.xlabel('a entry')
plt.ylabel('b entry')
plt.show()
```

```python
# Plotting Categorical Variable
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

fig, axes = plt.subplots(nrows=2, ncols=1)

axes[0].plot(t1, np.exp(t1), 'bo', t2, np.exp(t2), 'k')

axes[1].plot(t2, np.cos(2*np.pi*t2), 'r--')
```

```python
# Wroking with Text
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.5)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, 0.025, r'$\mu=100, \ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
