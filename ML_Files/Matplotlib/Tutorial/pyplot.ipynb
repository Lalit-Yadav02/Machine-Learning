```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])  # x and y are same so, x = 0-1 and y = 1-4
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) # plotting x vs y
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # plotting with red circle
```

```python
import numpy as np
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data) # data is like dictonary and other parma are key of data
plt.xlabel('a entry')
plt.ylabel('b entry')
plt.show()
```

```python
# Plotting Categorical Variable
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```
