```python
X = ecom_data.iloc[:, 3:-1]
y = ecom_data.iloc[:, -1:]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
linear_regressor = LinearRegression()
```

```python
linear_regressor.fit(X_train, y_train)
```

```python
linear_regressor.coef_
```

```python
prediction = linear_regressor.predict(X_test)
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.scatter(prediction, y_test)
axes.set_xlabel('Prdicted Data')
axes.set_ylabel('Test Data')
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(prediction, y_test)
```

```python
print('MAE', metrics.mean_absolute_error(prediction, y_test))
print('MSE ', metrics.mean_squared_error(prediction, y_test))
print('RMSE ', np.sqrt(metrics.mean_squared_error(prediction, y_test)))
```

```python
sns.distplot((y_test - prediction), bins = 50)
```
